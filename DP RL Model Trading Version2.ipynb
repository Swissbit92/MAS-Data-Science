### 5.1 Adding Custom Indicators to the data frame (Data-Processing)

```python
df = data

#adding a couple of indicators to the dataframe from the pandas_ta library:
#https://pypi.org/project/pandas-ta/
df['RSI'] = ta.rsi(df.Close, length=16)
df['OBV'] = TA.OBV(df)
df['ATR'] = TA.ATR(df)
df['SMA_20'] = TA.SMA(df, 20)
df['SMA_50'] = TA.SMA(df, 50)
#Bollinger Bands
df.ta.bbands(append=True)

#adding the MACD indicator to the data frame
# Calculate the 12-day and 26-day EMA
df_12EMA = df["Close"].ewm(span=12, adjust=False).mean()
df_26EMA = df["Close"].ewm(span=26, adjust=False).mean()

# Calculate the MACD and adding it to the data frame
df["MACD"] = df_12EMA - df_26EMA

#Adding some volatility, momentum and distance parameters to the data frame
df['Return'] = np.log(df['Close'] / df['Close'].shift(1))
df['Momentum'] = df['Return'].rolling(5).mean().shift(1)
df['Volatility'] = df['Return'].rolling(5).mean().shift(1)
df['Distance'] = (df['Close'] - df['Close'].rolling(50).mean()).shift(1)

#Adding a linear regression direction prediction to the data frame
#The 'Lags' are important / variable as they are needed to use for a prediction of the next forecast
lags = 5
cols = []
for lag in range(1, lags + 1):
    col = f'lag_{lag}'
    df[col] = df['Close'].shift(lag)
    cols.append(col)

df.dropna(inplace=True)
reg = np.linalg.lstsq(df[cols], df['Close'], rcond=None)[0]
df['Prediction'] = np.dot(df[cols], reg)

#clearing the data frame off from unwanted rows containing i.e. 'NaN' and checking afterwards with 'isna'
df.isna().sum()
```

```python
#Creating custom Signal features in order to define the training setup / 
#replacing the default indicators with the new custom indicators
def signals(env):
    #defining the the starting value based on frame_bound - starting at this first row
    start = env.frame_bound[0] - env.window_size
    #defining the ending indexes
    end = env.frame_bound[1]
    #grab the prices with the location indexer 'loc' and converting it into a numpy array
    prices = env.df.loc[:, 'Low'].to_numpy()[start:end]
    #defining which columns to be grabed for the environment 
    signal_features = env.df.loc[:, 
                                 ['Low', 'Volume', 'RSI', 'MACD', 
                                  'BBL_5_2.0', 'BBU_5_2.0', 'OBV',
                                 'ATR', 'SMA_20', 'SMA_50', 'Momentum',
                                  'Prediction', 'Return']].to_numpy()[start:end]
    return prices, signal_features
```

### 5.2 Create & Train new Env with the Custom Indicators

#### 5.4.1 A2C Model

Das A2C (Advantage Actor-Critic) ist ein Reinforcement-Learning-Algorithmus, der häufig in der KI-Forschung und im maschinellen Lernen verwendet wird, um agentenbasierte Probleme in der Umgebung des OpenAI Gym-Frameworks zu lösen. A2C kombiniert Konzepte des Policy-Gradienten-Verfahrens mit einem kritischen Bewertungsmodell.

Hier ist eine grundlegende Beschreibung des A2C-Algorithmus:

1. Initialisierung der Agenten: Es werden mehrere Agenten mit ihren eigenen neuronalen Netzwerken initialisiert. Jeder Agent ist dafür verantwortlich, Aktionen in der Umgebung zu wählen.

2. Erfassung von Daten: Jeder Agent interagiert mit der Umgebung, indem er Aktionen entsprechend seiner aktuellen Policy auswählt. Dabei werden Zustände, Aktionen, Belohnungen und andere Informationen gesammelt.

3. Berechnung des Vorteils: Basierend auf den gesammelten Daten wird der Vorteil jeder Aktion geschätzt. Der Vorteil gibt an, wie viel besser eine Aktion im Vergleich zu anderen Aktionen in einem bestimmten Zustand ist.

4. Aktualisierung des kritischen Bewertungsmodells: Das kritische Bewertungsmodell, auch Wertfunktion genannt, wird aktualisiert, um die Qualität der Aktionen zu bewerten. Dies kann beispielsweise durch den Vergleich der tatsächlichen Belohnungen mit den erwarteten Belohnungen erfolgen.

5. Aktualisierung der Policy: Die Gewichte der Policy werden basierend auf den gesammelten Daten und dem berechneten Vorteil aktualisiert. Dies geschieht in einem Schritt, der als "Advantage Actor-Critic" bezeichnet wird. Die Aktualisierung erfolgt durch Optimierungstechniken wie den stochastischen Gradientenabstieg.

6. Wiederholung der Schritte 2 bis 5: Dieser Prozess wird wiederholt, indem neue Daten gesammelt werden, das kritische Bewertungsmodell und die Policy aktualisiert werden, und die Agenten weiterhin mit der Umgebung interagieren.

Der A2C-Algorithmus zielt darauf ab, die Policy des Agenten schrittweise zu verbessern, indem er wiederholt Daten sammelt, das kritische Bewertungsmodell aktualisiert und die Policy aktualisiert. A2C nutzt die Vorteilsschätzung, um die Aktualisierung der Policy zu steuern und eine bessere Ausnutzung der Umgebung zu ermöglichen. Der Algorithmus wurde erfolgreich in verschiedenen Umgebungen des OpenAI Gym-Frameworks eingesetzt.

#### 5.4.2 PPO Model

Das PPO (Proximal Policy Optimization) ist ein Reinforcement-Learning-Algorithmus, der häufig in der KI-Forschung und im maschinellen Lernen verwendet wird, um agentenbasierte Probleme in der Umgebung des OpenAI Gym-Frameworks zu lösen.

PPO wurde entwickelt, um die Policy, also die Handlungsstrategie des Agenten, zu verbessern, indem es die Policy in kleinen Schritten aktualisiert, um eine gute Balance zwischen Exploration (Erkundung neuer Handlungsoptionen) und Exploitation (Nutzung bekannter guter Handlungen) zu finden. Der Algorithmus basiert auf einem Proximal Policy Optimization-Theorem, das eine Grenze für die Aktualisierungsschritte der Policy angibt, um übermäßige Änderungen zu vermeiden.

Hier ist eine grundlegende Beschreibung des PPO-Algorithmus:

1. Initialisierung der Policy: Die Policy wird mit zufälligen Gewichten initialisiert.

2. Erfassung von Daten: Der Agent interagiert mit der Umgebung, indem er Aktionen entsprechend seiner aktuellen Policy auswählt. Dabei werden Zustände, Aktionen, Belohnungen und andere Informationen gesammelt.

3. Berechnung von Vorteilen: Die gesammelten Daten werden verwendet, um die Vorteile der Aktionen zu schätzen. Die Vorteile zeigen an, wie gut eine Aktion im Vergleich zu anderen Aktionen in einem bestimmten Zustand ist.

4. Berechnung der Policy-Verlustfunktion: Basierend auf den gesammelten Daten wird eine Verlustfunktion erstellt, die angibt, wie stark die Policy aktualisiert werden soll. Bei PPO wird die Proximal Policy Optimization verwendet, um die Aktualisierungsschritte der Policy zu begrenzen.

5. Aktualisierung der Policy: Die Gewichte der Policy werden basierend auf der berechneten Verlustfunktion aktualisiert. PPO verwendet eine Optimierungstechnik wie den stochastischen Gradientenabstieg, um die Aktualisierung durchzuführen.

6. Wiederholung der Schritte 2 bis 5: Dieser Prozess wird wiederholt, indem neue Daten gesammelt werden, die Policy aktualisiert wird und der Agent weiterhin mit der Umgebung interagiert.

Der PPO-Algorithmus zielt darauf ab, die Policy des Agenten schrittweise zu verbessern, indem er wiederholt Daten sammelt, die Verlustfunktion berechnet und die Policy aktualisiert. Durch die Begrenzung der Aktualisierungsschritte wird PPO als stabil und effektiv angesehen und hat gute Ergebnisse in verschiedenen Umgebungen des OpenAI Gym-Frameworks erzielt.

```python
# Create a lof to save the path to to store the callback and model for later / so it does not need to be trained each time
log_dir = 'reinforement_learning/tmp_PPO/'
save_path = os.path.join('reinforcement_learning', 'tmp_PPO')
```

#### 5.4.3 DQN Model

Das DQN (Deep Q-Network) ist ein Reinforcement-Learning-Algorithmus, der entwickelt wurde, um agentenbasierte Probleme in der Umgebung des OpenAI Gym-Frameworks zu lösen. DQN verwendet tiefe neuronale Netzwerke, um eine Q-Funktion zu approximieren, die den Wert einer Aktion in einem gegebenen Zustand angibt.

Hier ist eine grundlegende Beschreibung des DQN-Algorithmus:

1. Initialisierung des Q-Netzwerks: Ein tiefes neuronales Netzwerk, das als Q-Netzwerk bezeichnet wird, wird initialisiert. Das Q-Netzwerk nimmt einen Zustand als Eingabe und gibt die Q-Werte für jede mögliche Aktion im Zustand aus.

2. Erfassung von Daten: Der Agent interagiert mit der Umgebung, indem er Aktionen basierend auf seiner aktuellen Policy auswählt. Dabei werden Zustände, Aktionen, Belohnungen, neue Zustände und weitere Informationen gesammelt. Diese Daten werden in einem sogenannten Erfahrungsspeicher gespeichert.

3. Erstellen des Trainingsziels: Für jeden Schritt im Erfahrungsspeicher wird ein Trainingsziel erstellt. Das Trainingsziel ist eine Schätzung des Q-Werts für die ausgewählte Aktion im aktuellen Zustand. Dazu werden die Q-Werte des Q-Netzwerks für den nächsten Zustand verwendet, gewichtet mit den zukünftigen Belohnungen.

4. Aktualisierung des Q-Netzwerks: Das Q-Netzwerk wird mit den Trainingszielen trainiert, um die Q-Werte zu verbessern. Dies erfolgt durch den Vergleich der geschätzten Q-Werte mit den tatsächlichen Q-Werten aus den Trainingszielen. Das Training erfolgt in einer Form des überwachten Lernens oder des Q-Lernens, bei dem die Gewichte des Q-Netzwerks angepasst werden.

5. Wiederholung der Schritte 2 bis 4: Dieser Prozess wird wiederholt, indem der Agent mit der Umgebung interagiert, Daten erfasst, Trainingsziele erstellt und das Q-Netzwerk aktualisiert werden.

Der DQN-Algorithmus zielt darauf ab, die Q-Funktion schrittweise zu verbessern, indem er wiederholt Daten sammelt, das Q-Netzwerk trainiert und die Aktionen des Agenten basierend auf den Q-Werten auswählt. DQN hat sich als effektiv erwiesen und wurde in verschiedenen Umgebungen des OpenAI Gym-Frameworks erfolgreich eingesetzt, insbesondere in Spielen wie Atari-Spielen.

```python
#env_with_indicators_testing = MyCustomEnv(df = df, window_size=20, frame_bound=(1200,1800)) 
runs2 = 100

env_with_indicators_testing2 = MyCustomEnv(df=df, window_size=window_size, frame_bound=(end_train_index, end_val_index))
env2 = MyCustomEnv(df=df, window_size=window_size, frame_bound=(end_train_index, end_val_index))
obs2 = env2.reset()
```

```python
buy_threshold = 0.15  # 15% increase in capital
sell_threshold = -0.05  # 5% decrease in capital
trade_duration_threshold = 10  # 10 days trade duration threshold

trade_duration = 0
is_trade_open = False
is_trade_pending = False

# Track the results
results = []
for run in range(runs2):
    obs2 = env2.reset()
    total_reward = 0
    total_profit = 0
    capital = 10000
    portfolio_value = [capital]

    while True:
        obs2 = obs2[np.newaxis, ...]
        # Using the trained model to predict what the future action should be to make a profit
        # The only two actions used here are 'buy' and 'sell'
        action, _states = model.predict(obs2)

        # Check if capital change meets the sell threshold
        if portfolio_value[-1] / capital >= 1 + buy_threshold:
            action = 1  # Sell action

        # Check if a trade is open and has exceeded the trade duration threshold
        if is_trade_open and trade_duration >= trade_duration_threshold:
            action = 1  # Sell action

        # Applying the actions to the environment based on the frame_bound
        obs2, rewards, done, info = env2.step(action)

        # Ensure rewards is a list
        if isinstance(rewards, int) or isinstance(rewards, float):
            rewards = [rewards]

        # Calculate the current portfolio value externally
        portfolio_value.append(portfolio_value[-1] + sum(rewards))

        # Update trade duration and trade status
        if action == 0:
            if not is_trade_pending and not is_trade_open:
                trade_duration = 0
                is_trade_pending = True
        else:
            if is_trade_pending:
                is_trade_pending = False
                is_trade_open = True
                capital = portfolio_value[-1]  # Set new capital as portfolio value
            else:
                trade_duration = 0
                is_trade_open = False

        # Increment trade duration if a trade is open
        if is_trade_open:
            trade_duration += 1

        total_reward += sum(rewards)
        total_profit = portfolio_value[-1] - capital

        if done:
            break

    results.append((portfolio_value[-1], total_reward, total_profit))

    print(f"Run: {run+1}")
    print(f"Final Portfolio Value: ${portfolio_value[-1]:.2f}")
    print(f"Total Reward: ${total_reward:.2f}")
    print(f"Total Profit: ${total_profit:.2f}")
    print("-------------------")

# Save the results for later visualization
np.save('portfolio_values.npy', np.array([result[0] for result in results]))
np.save('total_rewards.npy', np.array([result[1] for result in results]))
np.save('total_profits.npy', np.array([result[2] for result in results]))
```

## 7. Bewertung

Gängige Modelle und Ansätze, die häufig für den Handel verwendet werden:

Reinforcement Learning (RL): 
1. DQN (Deep Q-Network)
2. PPO (Proximal Policy Optimization)
3. A2C (Advantage Actor-Critic) 

können in OpenAI Gym eingesetzt werden, um Handelsstrategien zu entwickeln. Diese Modelle basieren auf der Optimierung der Entscheidungsstrategie des Handelsagenten basierend auf Rückkopplung und Belohnung aus der Umgebung.

Man sieht jedoch aus den Reports / Visualisierungen, dass dieser Ansatz auch unter Andwendung mehrerer TA Indikatoren, selten profitable ausfällt bzw. nicht verlässlich profitable ausfällt. Auch bei der Anwendung der Modelle auf die gleichen Zeitreihen ist eine starke Volatilität zu beobachten.

Es gibt mehrere Gründe warum der Ansatz von RL im Bezug auf timeseries oder noch spezifischer auf die Anwendung auf Trading Hindernisse mit sich bringt:

1. Hohe Dimensionalität: Der Aktienmarkt kann eine hohe Dimensionalität aufweisen, da es viele verschiedene Aktien, Indikatoren, Nachrichten und andere relevante Informationen gibt. Dies kann die Lernfähigkeit von RL-Algorithmen beeinträchtigen, da das Erlernen einer optimalen Handelsstrategie in einem solch großen Zustandsraum schwierig sein kann.

2. Stochastizität: Aktienmärkte sind oft durch stochastische Prozesse gekennzeichnet, bei denen zukünftige Preise und Marktbewegungen unvorhersehbar sind. Diese Unsicherheit kann es RL-Algorithmen erschweren, eine konsistente und profitable Handelsstrategie zu erlernen.

3. Geringe Verfügbarkeit von Daten: Im Vergleich zu anderen Anwendungsbereichen von RL wie z. B. Videospielen sind Finanzdaten begrenzt und historische Daten können nicht immer zuverlässige Vorhersagen für zukünftige Marktbewegungen liefern. Dies erschwert das Training eines RL-Modells, da es möglicherweise nicht genügend qualitativ hochwertige Daten gibt, um eine aussagekräftige Strategie zu entwickeln.

4. Nicht-Stationarität: Der Aktienmarkt ist ein sich ständig veränderndes System, in dem sich die Marktbedingungen, Trends und Volatilität im Laufe der Zeit ändern. Dies kann dazu führen, dass eine einmal erlernte Handelsstrategie schnell an Relevanz verliert und anpassungsfähige Modelle erfordert, die mit den sich ändernden Marktbedingungen umgehen können.

5. Manipulation und Rauschen: Die Finanzmärkte können anfällig für Manipulationen und Rauschen sein, wie z. B. unerwartete Ereignisse, falsche Informationen oder Marktmanipulationen. Solche Faktoren können die Effektivität von RL-Modellen beeinträchtigen, da sie Schwierigkeiten haben können, zwischen echten Signalen und Rauschen zu unterscheiden.

Obwohl RL für den Handel mit Aktien herausfordernd sein kann, werden dennoch verschiedene Ansätze erforscht und entwickelt, um diese Probleme zu überwinden. Dazu gehören Techniken wie die Verwendung von Ensemble-Modellen, die Integration von externem Wissen, die Verwendung von Deep Learning-Techniken und die Kombination von RL mit anderen Ansätzen wie Supervised Learning. Dennoch bleibt der Handel mit Aktien ein äußerst komplexes und unsicheres Umfeld, das weiterhin intensive Forschung erfordert, um effektive RL-Modelle zu entwickeln.
